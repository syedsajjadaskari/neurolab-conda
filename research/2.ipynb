```python
def get_historical_data(symbol, interval, start, end=None):
    bars = client.get_historical_klines(symbol=symbol, interval=interval, start_str=start, end_str=end, limit=1000)
    df = pd.DataFrame(bars)
    df["date"] = pd.to_datetime(df.iloc[:,0],unit="ms")
    df.columns = ["OPen Time", "Open", "High", "Low", "Close",
    "Volume", "Close Time","Quote Assest Volume", "Number of Trades",
    "Taker Buy Base Assest Volume", "Taker Buy Quote Volume", "Ignore", "Date"]
    df = df[["Date","Open","High", "Low","Close","Volume"]].copy()
    df.set_index("Date", inplace=True)
    for columns in df.columns:

        df[columns] = pd.to_numeric(df[columns], errors="coerce")
    return df
```
